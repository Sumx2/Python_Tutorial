Tuples



*   Ordered collection of items
*   Immutable



```python
### Creating a tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
```

```python
tpl=tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple = (1,"Sumanyu",9.24,True)
mixed_tuple
```

```python
### Accessing tuple elements
print(numbers[2])
print(numbers[-1])
print(numbers[2:4])
print(numbers[::3])
print(numbers[::-1])
```

```python
### Tuple operations

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
### Repeated values

mixed_tuple * 3
```

```python
numbers * 5
```

```python
### Immutable nature of tuples

lst = [1,2,3,4,5]
print(lst)
```

```python
lst[2]='Sumanyu'
```

```python
print(lst)
```

```python
numbers[1]='Sumanyu' ## immutable nature
```

```python
### Tuple methods

print(numbers.count(2))
```

```python
print(numbers.index(3)) ## returns first index of value
```

```python
### Packing and unpacking tuples

#packing
packed_tuple = 1,'Sumanyu',9.24
print(packed_tuple)
```

```python
#unpacking
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
### Unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
### NEsted tuples

## Nested list
lst=[['a','b'],[1,2],['Sumanyu',9.24]]
lst[0]
```

```python
lst[0][1]
```

```python
lst[1][0:2]
```

```python
lst=[['a','b'],[1,2],('Sumanyu',9.24)]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3),('a','b','c'),(True,False))
nested_tuple[1]
```

```python
nested_tuple[2][3]
```

```python
nested_tuple[1][::-1]
```

```python
### Iterating over nested tuples

for sub_tuple in nested_tuple:
  for item in sub_tuple:
    print(item,end="")
  print()
```
